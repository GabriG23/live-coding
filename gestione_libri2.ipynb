```python
class Book:
    def __init__(self,id=int, titolo=str, anno=int, autore=str) -> None:
        self.id = id
        self.titolo = titolo
        self.anno = anno
        self.autore = autore 

        def __str__(self):
            return f"ID: {self.id}, Titolo: {self.titolo}, Anno: {self.anno}, Autore: {self.autore}"
```

```python
#variabile che contiene dei libri. ogni libro ha il titolo, l'anno di publicazioen e l'autore
import time
import asyncio

lista_libri = [Book(1,"harry poter",2000,"j.k rowling"), Book(2,"game of throne",2025,"j.r.r martins"), Book(3,"animali fantastici",1980,"j.k rowling")]
    
#print(lista_libri[1]["titolo"])

#creare una funzione che stampa i libri
def stampa_libri(lista):
    for libro in lista:
       print(libro)

#funzione che aggiunge nuovo libro, con id automatico
def add_book(titolo:str,anno:int,autore:str):
    flag=False #non c'è il libro
    for libro in lista_libri:
        if libro.titolo == titolo: 
           flag=True
    
    if flag == True:
       print(f"{titolo} gia presente nel database")
    else:    
        indice=len(lista_libri)+1
        print("inserimento libro in corso...")
        print(indice)
        voce = Book(indice,titolo,anno,autore)
        lista_libri.append(voce)
        print("libro inserito con successo.")

#funzione async di add_book
async def add_book_async(titolo:str,anno:int,autore:str):
    flag=False #non c'è il libro
    for libro in lista_libri:
        if libro.titolo == titolo: 
            flag=True

    if flag == True:
        print(f"{titolo} gia presente nel database")
    else:    
        indice=len(lista_libri)+1
        print("inserimento libro in corso...")
        await asyncio.sleep(3)
        print(indice)
        voce = Book(indice,titolo,anno,autore)
        lista_libri.append(voce)
        print("libro inserito con successo.")
            
#funzione per rimuovere un libro
def remove_libro(titolo):
     for libro in lista_libri:
          if libro.titolo == titolo:
               lista_libri.remove(libro)
               print("libro cancellato")

#crea una funzione che ordini i libri dal più piccolo al più grande
def ordinaAnno(lista_lib):
     print("Ordinamento in corso...")
     
     lista_lib = lista_libri
     lista_lib = sorted(lista_libri, key= lambda a: a.anno)
     return lista_lib
```
